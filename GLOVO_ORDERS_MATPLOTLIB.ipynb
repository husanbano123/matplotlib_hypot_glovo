# One sample
the avg rating is above 3.0.(note the given data is of sample  where mean=2.9 which is sample mean)
H0=mu(mean of population)<=3.0
H1=mu>3.0

```python
ratings=df['Rating']
t_stats,p_value=stats.ttest_1samp(ratings,popmean=3.0)
alpha=0.05
```

```python
p_value/2>alpha
```

```python
t_stats
```

```python
p_value/2
```

```python
if p_value/2<alpha:
    print("Rejecting Null Hypothesis suggesting avg ratings is sifnifically more than 3.0")
else:
    print("Fail to Reject Null Hypothesis H0 suggesting avg ratings is sifnifically les than 3.0")
print("Accept alternate Hypoyhesis H1")
```

# independant sample
2.Two tailed test:
comparing mean of 2 independant samples i.e "rating" of premium  vs regular
H0= no significant difference in rating between regular and premium
H1= there is significance difference in rating between regular and premium

```python
regular_ratings=df[df['Membership_Type']=="Regular"]["Rating"]
premium_ratings=df[df['Membership_Type']=="Premium"]["Rating"]
```

```python
regular_ratings
```

```python
premium_ratings
```

```python
t_stats1,p_val1=stats.ttest_ind(regular_ratings,premium_ratings)
print("\nTwo_tailed test results:")
print("T_statistics:",t_stats1)
print("P_value: ",p_val1)
alpha=0.05
```

```python
p_val1>alpha
```

```python
if p_val1<alpha:
    print("Reject null hypothesis H0,suggesting there is significant difference in rating between regular and premium")
else:
    print("Fail to reject null hypo H0,suggesting there is no significant difference in rating between regular and premium")
```

#  Anova sample
1.ONE WAY ANOVA

whether there are any statistically significant difference between the means of multiple groups(Gender in this case)

H0=no significant difference in means among the Gender in terms of Age
H1= at least one dish has a different mean in terms of Age

```python
female_age=df[df["Gender"]=="Female"]["Age"]
male_age=df[df["Gender"]=="Male"]["Age"]
other_age=df[df["Gender"]=="Other"]["Age"]
```

```python
f_stat,p_val=stats.f_oneway(female_age,male_age,other_age)
print("\nANOVA Results:")
print("F_statistics:",f_stat)
print("P_value: ",p_val)
alpha=0.05
```

```python
p_val>alpha
```

```python
if p_val<alpha:
    print("Reject null hypothesis ,suggesting at least one Gender has significant difference in Age between male,female and others")
else:
    print("Fail to Reject null hypothesis H0,suggesting there is no significant difference in Age between Genders")
```
